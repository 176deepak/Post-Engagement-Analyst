```python
# @title Import the dependencies/modulues
import os
import shutil
import random
import numpy as np
import pandas as pd
import kagglehub
```

```python
# @title Download the latest version Data
path = kagglehub.dataset_download("propriyam/instagram-data")
shutil.move(path, '/content/data')
print("Path to dataset files:", '/content/data')
```

```python
# @title Read the data
df = pd.read_csv('/content/data/instagram_data.csv')
df.sample(n=10)
```

```python
# info
df.info()
```

```python
# @title Drop the columns, Which is not usefull
df.drop(columns=['shortcode', 'created_at', 'imageUrl', 'username'], inplace=True)
df.sample(5)
```

```python
# @title Convert type str to bool
mapper = {'True': True, 'False': False}

# first check by value
if type(df['is_video'][0]) != bool:
    df['is_video'] = df['is_video'].map(mapper)

if type(df['multiple_images'][0]) != bool:
    df['multiple_images'] = df['multiple_images'].map(mapper)

df['likes'] = pd.to_numeric(df['likes'], errors='coerce')
df['comments'] = pd.to_numeric(df['comments'], errors='coerce')

df.sample(2)
```

```python
# @title Post Type feature

# create new column "post_type" based on "is_video" & "multiple_images"

df['post_type'] = np.select(
    [
        (df['is_video'] == True),
        (df['is_video'] == False) & (df['multiple_images'] == False),
        (df['is_video'] == False) & (df['multiple_images'] == True)
    ],
    ['reel', 'static image', 'carousel'],
    default='unknown'
)
```

```python
# @title Redefine location

def extract_loc(x):
    if isinstance(x, str):
        try:
            x = eval(x)
            loc = x.get('name', None)
            return loc
        except (SyntaxError, NameError, TypeError):
            pass
    return x

df['location'] = df['location'].apply(extract_loc)
df.sample(3)
```

```python
# @title Drop "is_video" & "multiple_images"

df.drop(columns=['is_video', 'multiple_images'], inplace=True)
df.sample(2)
```

```python
# @title Share feature

# Define the rule to calculate shares
def calculate_shares(row):
    comments = row['comments']
    likes = row['likes']

    # Base shares on comments with random adjustment
    if np.isnan(comments) or np.isnan(likes):
        return np.nan

    shares = comments + np.random.randint(0, max(1, int(comments * random.uniform(0.3, 0.8))))

    if likes > 0:
        shares = min(shares, int(likes * 0.8))

    return shares

df['shares'] = df.apply(calculate_shares, axis=1)
df.sample(3)
```

```python
# Fill na values by empty string
df['owner_username'] = df['owner_username'].fillna('')
df['caption'] = df['caption'].fillna('')
df['comments'] = df['comments'].fillna('')
df['likes'] = df['likes'].fillna('')
df['location'] = df['location'].fillna('')
df['followers'] = df['followers'].fillna('')
df['following'] = df['following'].fillna('')
df['shares'] = df['shares'].fillna('')
df['post_type'] = df['post_type'].fillna('')
df.sample(3)
```

```python
# Save the csv file
df.to_csv('instagram.csv', index=False)
```
